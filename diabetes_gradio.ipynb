```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import gradio as gr
```

```python
diabetes = pd.read_excel("/content/logisticdiabetes(1).xlsx")
```

```python
X = diabetes[['plas', 'pres', 'insu', 'mass']]
y = diabetes['class']
```

```python
model = LogisticRegression()
model.fit(X, y)
```

```python
def predict_diabetes(plas, pres, insu, mass):
    input_data = np.array([[plas, pres, insu, mass]])
    prediction = model.predict(input_data)[0]
    return f"Prediction: {prediction.replace('_', ' ').capitalize()}"
```

```python
interface = gr.Interface(
    fn=predict_diabetes,
    inputs=[
        gr.Number(label="Plasma Glucose (plas)"),
        gr.Number(label="Blood Pressure (pres)"),
        gr.Number(label="Insulin Level (insu)"),
        gr.Number(label="Body Mass Index (mass)")
    ],
    outputs="text",
    title="Diabetes Prediction App",
    description="Enter medical values to predict whether the person is diabetic using Logistic Regression model."
)
```

```python
interface.launch(share=True)
```
