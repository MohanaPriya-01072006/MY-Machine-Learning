```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
data = pd.read_csv('/content/Ex 3.csv')
print("Dataset preview:")
print(data.head())

# Define features and target
x = data[["age", "income"]]
y = data["bought"]
print(x)

from sklearn.model_selection import train_test_split
xtr, xte, ytr, yte = train_test_split(x, y, test_size=1/3, random_state=0)

# Import SVC and create the model
from sklearn.svm import SVC
svm_model = SVC(kernel='linear', probability=True)  # probability=True if you want predict_proba later

# Train the SVM model
svm_model.fit(xtr, ytr)

# Make predictions on test and train data
tepr = svm_model.predict(xte)
trpr = svm_model.predict(xtr)

from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

print("Predictions on test set:", tepr.tolist())

# Calculate accuracy
accuracy = accuracy_score(yte, tepr)
print("Accuracy:", accuracy)

# Compute and display confusion matrix
cm = confusion_matrix(yte, tepr)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Not Bought (0)", "Bought (1)"])
disp.plot(cmap=plt.cm.Blues)
plt.title(f"Confusion Matrix (Accuracy={accuracy:.2f})")
plt.show()
```
